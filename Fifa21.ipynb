Include the list of Position in the data. (Like a OneHotEncoder) 

```python
clean= nums[['Age','Growth','Attacking','Skill','Movement','Power','Mentality','Defending','Goalkeeping','Base Stats','OVA']]
poslist=['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM', 'RAM',
       'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB',
       'LCB', 'CB', 'RCB', 'RB', 'GK']

for x in poslist:
       clean[x]=0.

clean=pd.concat([clean,chars['Position']],axis=1)
clean.head()       
```

```python
def pos_encoder(row,poslist):
    pos=row['Position'].split()

    for x in pos:
        if poslist.count(x)!=0:
            row[x]=1.
    return row

#chars['Position'].head(5).isna().sum()
chars.shape
#print(clean.head())
clean=clean.apply(lambda row: pos_encoder(row,poslist),axis=1)
clean=clean.drop(['Position'],axis=1)
clean.head()
```

```python
y=clean['OVA']
X=clean.drop(['OVA'],axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
predictions = lm.predict(X_train)
r2=r2_score(y_train, predictions)
r2
```
